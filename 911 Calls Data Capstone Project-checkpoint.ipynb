```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('911.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['zip'].value_counts().head()
```

```python
df['twp'].value_counts().head()
```

```python
df['title'].nunique()
```

```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
```

```python
df['Reason'].value_counts().head(1)
```

```python
sns.countplot('Reason', data=df)
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['timeStamp']= pd.to_datetime(df['timeStamp'])
```

```python
df['Hour']= df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
df['Day of Week']
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df.head()
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
df['Day of Week']
```

```python
sns.countplot(x='Day of Week', data= df, hue='Reason')
```

```python
sns.countplot(x='Month', data=df, hue='Reason', palette='inferno')
```

```python
byMonth = df.groupby('Month').count()
byMonth.head()
```

```python
sns.set_style('darkgrid')
byMonth['twp'].plot()
```

```python
sns.lmplot(x='Month', y='twp', data=byMonth.reset_index())
```

```python
df['Date']= df['timeStamp'].apply(lambda time: time.date())
df['Date']
```

```python
byDate = df.groupby('Date').count()['twp'].plot()
plt.tight_layout()
```

```python
df[df['Reason']=='Traffic'].groupby('Date').count()['lat'].plot()

plt.tight_layout()
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['lat'].plot()

plt.tight_layout()
```

```python
df[df['Reason']=='EMS'].groupby('Date').count()['lat'].plot()

plt.tight_layout()
```

```python
df1 = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
```

```python
df1
```

```python
sns.heatmap(df1)
```

```python
sns.clustermap(df1, cmap='mako')
```
