```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('kyphosis.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
sns.set_style('whitegrid')
sns.pairplot(df, hue='Kyphosis', diag_kind='hist')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Kyphosis', axis=1)
y=df['Kyphosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

```python
predic= dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(predic, y_test))
print('\n')
print(confusion_matrix(predic, y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X_train, y_train)
```

```python
rfcpred= rfc.predict(X_test)
```

```python
print(confusion_matrix(rfcpred, y_test))
print('/n')
print(classification_report(rfcpred, y_test))
```
